Aprendizagem por Reforço Profundo é uma técnica de Inteligência Artificial para treinamento de modelos para tomada de uma sequência de decisões. O modelo deve ser capaz de atingir a melhor sequência de decisões a partir do treinamento que consiste em obter o máximo de recompensas. Este sistema de recompensas faz com que, durante o treinamento, o modelo tome decisões baseadas em tentativa e erro, onde cada erro cause uma penalização e o acerto gere uma recompensa, fazendo com que a sequência que produza a maior recompensa seja a melhor. Nesta técnica, o modelo não recebe informações de soluções, tornando-o autônomo nas suas soluções, dando a ele uma capacidade imensa de criatividade, muito maior que as dos seres humanas, quanto maior for a pesquisa e quantidade de tentativas durante o aprendizado.

O objetivo do jogo é equilibrar o poste sobre um carrinho. #Questão 5
A lógica Fuzzy, ou lógica difusa, é uma lógica multivalorada que possui como objetivo modelar raciocínios aproximados, ou seja, incertos. Diferentemente da lógica Booleana que lida com certezas, baseando-se em verdadeiro e falso, a lógica Fuzzy lida com parcialidades, onde pode assumir qualquer valor entre 0 e 1, significando que assumir valores verdadeiros, falsos e parcialmente verdadeiro ou falso. Esses valores de verdade são representados por funções de pertinência, que atribuem um grau de pertinência a cada valor em um conjunto universo. Essas funções de pertinência são geralmente representadas por curvas suaves, e podem refletir a incerteza ou imprecisão associada a um determinado valor. Como a lógica está associada à conjuntos, é importante o uso dos operadores NOT, AND e OR, onde os dois últimos no contexto de Fuzzy são conhecidos respectivamente como mínimo e máximo; esses operadores Fuzzy permitem realizar inferências e tomar decisões baseadas em informações imprecisas ou vagas. Num exemplo de aplicação em ar-condicionado, podem ser utilizadas regras de resfriamento e aquecimento com objetivo de diminuir o custo da utilização do aparelho controlando o consumo de energia através da potência usada em determinadas temperaturas, fazendo uso de sensores para enviar informações de temperatura do ambiente para serem empregadas regras Fuzzy para que o objetivo seja alcançado. Outro exemplo é o uso de controladores Fuzzy em aspirador de pó para ajustar a potência sucção para determinadas superfícies e/ou quantidade de sujeira que queira limpar, economizando energia.

#Questão 6
Uma árvore é uma estrutura de dados não linear que consiste em um conjunto de nós interconectados de forma hierárquica. Cada nó em uma árvore é chamado de nó pai e pode ter zero ou mais nós filhos, que são conectados a ele por arestas. O nó no topo da árvore é chamado de raiz, e os nós que não possuem filhos são chamados de nós folha ou nós terminais. Algumas árvores comuns são Árvores Binárias, Balanceadas e Ternárias. Alguns métodos de busca em árvore são Busca em largura (BFS - Breadth-First Search), Busca em profundidade (DFS - Depth-First Search), Busca de custo uniforme. Por exemplo, o algoritmo A-estrela é uma combinação de busca por largura e busca de custo uniforme que pode ser utilizado tanto em árvores quanto em grafos. Sua principal função é encontrar o menor caminho, sendo muito usado em inteligência artificial. Uma aplicação neste sentido é para cálculo da menor rota em um mapa, onde o mapa será representado como uma estrutura de árvore, onde os nós representam as interseções no mapa e as arestas representam as ruas que conectam essas interseções. Usando o algoritmo, pode-se atribuir uma função de avaliação (f(n)) para cada nó no mapa. Essa função de avaliação é a soma do custo acumulado (g(n)), que é o comprimento do caminho percorrido até o nó atual, e uma heurística (h(n)), que estima a distância direta (ou distância de Manhattan) do nó atual até o ponto de chegada. Durante a busca, o algoritmo explorará os nós da árvore em uma ordem que priorize aqueles com a menor função de avaliação f(n). Ele usará a heurística para tentar direcionar a busca na direção correta, buscando os caminhos mais promissores em direção ao ponto de chegada. Conforme o A-estrela explora os nós, ele expande e avalia seus nós filhos, considerando o custo acumulado até o momento e a heurística para cada nó, avançando pela árvore até encontrar o ponto de chegada ou até explorar todos os caminhos possíveis.
Possível código para este algoritmo:

```python
import heapq

class Node:
    def __init__(self, name):
        self.name = name
        self.neighbors = []
        self.g = float('inf')  # Custo acumulado do nó inicial até este nó
        self.h = 0  # Heurística: estimativa da distância até o objetivo
        self.f = float('inf')  # Função de avaliação f(n) = g(n) + h(n)
        self.parent = None

    def add_neighbor(self, neighbor, cost):
        self.neighbors.append((neighbor, cost))

    def __lt__(self, other):
        return self.f < other.f

def astar(start, goal):
    # Inicialização
    start.g = 0
    start.f = start.h

    open_list = []
    closed_list = set()
    heapq.heappush(open_list, start)

    while open_list:
        current = heapq.heappop(open_list)

        if current == goal:
            # Encontrou o objetivo, reconstrói o caminho
            path = []
            while current:
                path.append(current.name)
                current = current.parent
            path.reverse()
            return path

        closed_list.add(current)

        for neighbor, cost in current.neighbors:
            if neighbor in closed_list:
                continue

            g = current.g + cost
            if g < neighbor.g:
                neighbor.g = g
                neighbor.h = heuristic(neighbor, goal)  # Calcula a heurística
                neighbor.f = g + neighbor.h
                neighbor.parent = current

                if neighbor not in open_list:
                    heapq.heappush(open_list, neighbor)

    # Não foi encontrado um caminho
    return None

def heuristic(node, goal):
    # Heurística simples: distância em linha reta entre os nós
    x1, y1 = node.name
    x2, y2 = goal.name
    return abs(x2 - x1) + abs(y2 - y1)

# Criando o mapa de exemplo
A = Node((0, 0))
B = Node((1, 0))
C = Node((1, 1))
D = Node((0, 1))
E = Node((2, 1))
F = Node((2, 2))
G = Node((1, 2))
H = Node((0, 2))
I = Node((3, 1))

A.add_neighbor(B, 1)
B.add_neighbor(A, 1)
B.add_neighbor(C, 1)
B.add_neighbor(D, 1)
C.add_neighbor(B, 1)
C.add_neighbor(F, 3)
D.add_neighbor(B, 1)
D.add_neighbor(E, 2)
D.add_neighbor(G, 1)
E.add_neighbor(D, 2)
E.add_neighbor(F, 1)
F.add_neighbor(C, 3)
F.add_neighbor(E, 1)
F.add_neighbor(I, 3)
G.add_neighbor(D, 1)
G.add_neighbor(H, 1)
H.add_neighbor(G, 1)
H.add_neighbor(I, 2)
I.add_neighbor(F, 3)
I.add_neighbor(H, 2)

# Executando a busca A*
start = A
goal = I
path = astar(start, goal)
print(f"Caminho entre os nós A e I:")
if path:
    print("Caminho encontrado:", path)
else:
    print("Caminho não encontrado.")

start = B
goal = F
path = astar(start, goal)
print(f"Caminho entre os nós B e F:")
if path:
    print("Caminho encontrado:", path)
else:
    print("Caminho não encontrado.")
```

Trabalho sobre agentes inteligentes em problemas de tomadas de decisões, cujo problema é jogar [Connect Four](https://en.wikipedia.org/wiki/Connect_Four). Este jogo é composto por um quadro e fichas de duas cores que serão divididas entre os dois jogadores, cada jogador insere a ficha em uma coluna do quadro com o objetivo de formar uma linha vertical, horizontal ou diagonal de fichas, o vencedor será o jogador que formar esta linha primeiro com suas respectivas fichas.

Nesta solução é utilizada a abordagem de treinamento dos agentes através do aprendizado por reforço, não utilizando heurística para fazer o treinamento dos agentes inteligentes. Com essa estratégia os agentes serão treinados puramente pela quantidade de vezes que jogarem, buscando atingir a maior taxa de vitórias para refinar sua estratégia. Inicialmente os pesos são atribuídos de forma aleatória, durante o treinamento passam a serem definidos pela a recompensa por cada movimento e a soma acumulativa das recompensas, visando, em média, maximizar a recompensa do agente, assim atualizando os pesos a cada partida.

O código a seguir foi retirado do tutorial [Deep Reinforcement Learning
](https://www.kaggle.com/code/alexisbcook/deep-reinforcement-learning), feito pela instrutora Alexis Cook e disponibilizado na plataforma [Kaggle](https://www.kaggle.com/), muito utilizada para compartilhamento de conhecimento em Inteligência Artificial.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import gym
from kaggle_environments import make, evaluate
from gym import spaces

class ConnectFourGym(gym.Env):
    def __init__(self, agent2="random"):
        ks_env = make("connectx", debug=True)
        self.env = ks_env.train([None, agent2])
        self.rows = ks_env.configuration.rows
        self.columns = ks_env.configuration.columns
        # Learn about spaces here: http://gym.openai.com/docs/#spaces
        self.action_space = spaces.Discrete(self.columns)
        self.observation_space = spaces.Box(low=0, high=2,
                                            shape=(1,self.rows,self.columns), dtype=int)
        # Tuple corresponding to the min and max possible rewards
        self.reward_range = (-10, 1)
        # StableBaselines throws error if these are not defined
        self.spec = None
        self.metadata = None
    def reset(self):
        self.obs = self.env.reset()
        return np.array(self.obs['board']).reshape(1,self.rows,self.columns)
    def change_reward(self, old_reward, done):
        if old_reward == 1: # The agent won the game
            return 1
        elif done: # The opponent won the game
            return -1
        else: # Reward 1/42
            return 1/(self.rows*self.columns)
    def step(self, action):
        # Check if agent's move is valid
        is_valid = (self.obs['board'][int(action)] == 0)
        if is_valid: # Play the move
            self.obs, old_reward, done, _ = self.env.step(int(action))
            reward = self.change_reward(old_reward, done)
        else: # End the game and penalize agent
            reward, done, _ = -10, True, {}
        return np.array(self.obs['board']).reshape(1,self.rows,self.columns), reward, done, _
```

```python
# Create ConnectFour environment
env = ConnectFourGym(agent2="random")
```

```python
import torch as th
import torch.nn as nn


from stable_baselines3 import PPO
from stable_baselines3.common.torch_layers import BaseFeaturesExtractor

# Neural network for predicting action values
class CustomCNN(BaseFeaturesExtractor):

    def __init__(self, observation_space: gym.spaces.Box, features_dim: int=128):
        super(CustomCNN, self).__init__(observation_space, features_dim)
        # CxHxW images (channels first)
        n_input_channels = observation_space.shape[0]
        self.cnn = nn.Sequential(
            nn.Conv2d(n_input_channels, 32, kernel_size=3, stride=1, padding=0),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=0),
            nn.ReLU(),
            nn.Flatten(),
        )

        # Compute shape by doing one forward pass
        with th.no_grad():
            n_flatten = self.cnn(
                th.as_tensor(observation_space.sample()[None]).float()
            ).shape[1]

        self.linear = nn.Sequential(nn.Linear(n_flatten, features_dim), nn.ReLU())

    def forward(self, observations: th.Tensor) -> th.Tensor:
        return self.linear(self.cnn(observations))

policy_kwargs = dict(
    features_extractor_class=CustomCNN,
)

# Initialize agent
model = PPO("CnnPolicy", env, policy_kwargs=policy_kwargs, verbose=0)
```

```python
# Train agent
model.learn(total_timesteps=60000)
```

```python
def agent1(obs, config):
    # Use the best model to select a column
    col, _ = model.predict(np.array(obs['board']).reshape(1, 6, 7))
    # Check if selected column is valid
    is_valid = (obs['board'][int(col)] == 0)
    # If not valid, select random move.
    if is_valid:
        return int(col)
    else:
        return random.choice([col for col in range(config.columns) if obs.board[int(col)] == 0])
```

```python
def get_win_percentages(agent1, agent2, n_rounds=100):
    # Use default Connect Four setup
    config = {'rows': 6, 'columns': 7, 'inarow': 4}
    # Agent 1 goes first (roughly) half the time
    outcomes = evaluate("connectx", [agent1, agent2], config, [], n_rounds//2)
    # Agent 2 goes first (roughly) half the time
    outcomes += [[b,a] for [a,b] in evaluate("connectx", [agent2, agent1], config, [], n_rounds-n_rounds//2)]
    print("Agent 1 Win Percentage:", np.round(outcomes.count([1,-1])/len(outcomes), 2))
    print("Agent 2 Win Percentage:", np.round(outcomes.count([-1,1])/len(outcomes), 2))
    print("Number of Invalid Plays by Agent 1:", outcomes.count([None, 0]))
    print("Number of Invalid Plays by Agent 2:", outcomes.count([0, None]))
```
